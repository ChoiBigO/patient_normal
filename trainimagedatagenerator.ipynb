```
#하나의 이미지당 2개의 이미지를 생성하여 imagegenerator 폴더에 저장하는 코드
#문제 없는 코드

import numpy as np

# 랜덤시드 고정시키기
np.random.seed(0)

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

# 데이터셋 불러오기 및 이미지 생성 조건
# 아래 코드와는 다르게 이미지의 변화가 크게 발생하지 않도록 옵션 변경
train_datagen = ImageDataGenerator(rescale=1./255, 
                                  rotation_range=0,
                                  width_shift_range=0,
                                  height_shift_range=0,
                                  shear_range=0,                              
                                  horizontal_flip=True,
                                  vertical_flip=0,
                                  fill_mode='nearest')

#uniform 데이터 생성                                   
#총 이미지는 100개 
for j in range(100):
  #변형시킬 이미지 지정
  img = load_img('/content/gdrive/My Drive/resnet1/train/uniform/'+str(j+1)+'.jpg')
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)
  i = 0
  #이미지 저장할 주소공간 지정
  #이미지 이름의 시작은 j + 101, 저장 포맷은 jpg
  for batch in train_datagen.flow(x, 
                                  batch_size=1, 
                                  save_to_dir='/content/gdrive/My Drive/resnet1/imagegenerator2/uniform',  
                                  save_prefix= str(j+101), 
                                  save_format='jpg'):
    i += 1
    if i > 1: 
        break

#casual 데이터 생성   
for k in range(100):
  img = load_img('/content/gdrive/My Drive/resnet1/train/casual/'+str(k+1)+'.jpg')
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)
  i = 0
  for batch in train_datagen.flow(x, 
                                  batch_size=1, 
                                  save_to_dir='/content/gdrive/My Drive/resnet1/imagegenerator2/casual',  
                                  save_prefix= str(k+101), 
                                  save_format='jpg'):
    i += 1
    if i > 1: 
        break
```
